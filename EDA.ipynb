```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentPerformanceFactors.csv')
df
```

```python
study_cols = [
    "Hours_Studied",
    "Attendance",
    "Sleep_Hours",
    "Motivation_Level",
    "Parental_Involvement",
    "Access_to_Resources",
    "Tutoring_Sessions",
    "Internet_Access",
    "Extracurricular_Activities",
    "Peer_Influence",
    "Learning_Disabilities",
    "Physical_Activity",
    "Gender"
]
```

```python
eda_df = df[study_cols].copy()
eda_df
eda_df.to_csv("FilterColumns.csv", index=False)
```

```python
print("Shape of dataset:", eda_df.shape)
print("\nColumn info:")
print(eda_df.info())
```

```python
print("\nFirst 5 rows:")
print(eda_df.head())
```

```python
print("\nMissing values per column:")
print(eda_df.isnull().sum())
```

```python
print("\nSummary statistics (numeric):")
print(eda_df.describe())
```

```python
print("\nSummary statistics (categorical):")
print(eda_df.describe(include="object"))
```

```python
# Check for duplicate rows
duplicate_rows = eda_df[eda_df.duplicated()]
print("\nNumber of duplicate rows:", duplicate_rows.shape[0])
```

```python
numeric_cols = [
    "Hours_Studied", "Attendance", "Sleep_Hours",
    "Tutoring_Sessions", "Physical_Activity"
]

cat_cols = [
    "Motivation_Level", "Parental_Involvement", "Access_to_Resources",
    "Internet_Access", "Extracurricular_Activities", "Peer_Influence",
    "Learning_Disabilities", "Gender"
]
```

```python
eda_df[numeric_cols].hist(bins=20, figsize=(12, 8), color="#87CEEB", edgecolor="black")
plt.suptitle("Distribution of Numeric Study Habit Features", fontsize=14, y=1.02)
plt.tight_layout()
plt.show()
```

```python
for col in cat_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=eda_df, x=col)
    plt.title(f"Count of {col}", fontsize=13)
    plt.xticks(rotation=20)
    plt.tight_layout()
    plt.show()
```

```python
plt.figure(figsize=(8, 6))
corr = eda_df[numeric_cols].corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Correlation Heatmap of Study Habit Factors", fontsize=13)
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(
    eda_df,
    vars=["Hours_Studied", "Sleep_Hours", "Physical_Activity"],
    hue="Gender",
    diag_kind="hist"
)
plt.suptitle("Pairwise Relationships Between Key Study Variables", fontsize=14, y=1.02)
plt.show()
```

```python
print("\n Key Insights from EDA:")
print("- Dataset is clean with no missing values in key study-related features.")
print("- Most students study ~20 hours/week and maintain ~80% attendance.")
print("- Average sleep duration is ~7 hours â€” balanced for learning efficiency.")
print("- Majority have medium motivation and parental involvement levels.")
print("- Over 90% have internet access, enabling digital learning habits.")
print("- Extracurricular activities participation is high (~60%), indicating balanced habits.")
print("- Physical activity averages around 3 hours/week â€” promoting mental balance.")
print("- Gender distribution is balanced; no major skew.")
print("- These patterns can guide K-Means clustering into behavioral groups like 'Focused', 'Balanced', and 'Low Engagement'.")
```
